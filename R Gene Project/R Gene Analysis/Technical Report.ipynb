### Quality Control:
I first started by downloading the rcc files using the nanostringr library. From there I extracted just the data portion of the file and put it into a table ([Table 1: Data](#data)). Then I started to set the data table up for normalization by extracting the string values ([see Table 2: New Data](#ndata)).  I then calculated the normal factors and applied the TMM method for normalization for each gene, which was saved as a new table ([see Table 3: Normalized Data](#allNormData)). After that, I made a table of positive and negative genes using table 1 ([see Table 4: Positive/Negative Data](#PNdata)) to get names to filter table 3. With the names I made the filtered normalized data for the heat map ([see Table 5: Normalized Positive/Negative Data](#NormPN)). I then applied clustering to table 5 and used it to make the heat map([see Heat Map](#heatmap)). 
### Data Analysis:
To start with building the box plots I filtered the genes from table 3 to make a new table ([see Table 6: Box Plot Data](#BPdata)).  I then separated the two genes into their own tables ([see Table 7: MCL1 Unsorted](#MCL1), [Table 8: CXCL1 Unsorted](#CXCL1)). From there I started to work on preparing table 7 data for the box plot by separating the two timepoints into their own tables ([see Table 9: MCL1 Base](#MCL1Base), [Table 10: MCL1 Post](#MCL1Post)). After that I combined table 9 and 10 to make a table that contains both timepoints ([see Table 11: MCL1 Sorted](#BoxMCL1)). Once I was done preparing the data for the MCL1 gene, I repeated the same process for the CXCL1 gene, creating three new tables ([see Table 12: CXCL1 Base](#CXCL1Base), [Table 13: CXCL Post](#CXCL1Post), [Table 14: CXCL Sorted](#BoxCXCL1)). After that I calculate the means for each column in table 12 and 14. With my data ready I entered table 11 and 14 into the boxplot function to create the boxplot for both genes ([see Box Plot](#boxplot)). 


```python
import glob
import pandas as pd
import os
csvs=[]
path = os.getcwd()
csv_files = sorted(glob.glob(os.path.join(path, "*.csv")),key=os.path.getmtime)
for file_name in csv_files:
    df=pd.read_csv(file_name)
    csvs.append(df)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```
